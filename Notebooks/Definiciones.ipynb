## Definiciones de conceptos básicos

Primero veremos unas definiciones fundamentales para las series de tiempo y ARIMA.

### 1. Serie de Tiempo

Una serie de tiempo es una realización de un **proceso estocástico**. Se define como una sucesión de variables aleatorias $Z_t$ indexadas en el tiempo, donde $t$ pertenece a un conjunto de índices $T$ (generalmente discretos, $t = 1, 2, ..., n$).

Formalmente, un proceso estocástico es una familia de variables aleatorias:
$$\{Z_t : t \in T\}$$


### 2. Serie Estacionaria

La estacionariedad es el supuesto fundamental para la modelación ARIMA. Una serie se considera **estacionaria en sentido débil (o amplio)** si sus propiedades estadísticas básicas son invariantes a través del tiempo. Esto implica que el comportamiento del proceso se mantiene estable en torno a una media constante y con una varianza finita.

Matemáticamente, un proceso $\{Z_t\}$ es débilmente estacionario si cumple las siguientes tres condiciones para todo $t$:

1.  **Media Constante:** La esperanza matemática no depende del tiempo.
    $$E[Z_t] = \mu < \infty$$
    
2.  **Varianza Constante:** La dispersión de los datos alrededor de la media es finita y constante.
    $$Var(Z_t) = E[(Z_t - \mu)^2] = \sigma_Z^2 < \infty$$
    
3.  **Autocovarianza dependiente solo del retardo:** La relación entre dos observaciones $Z_t$ y $Z_{t+k}$ depende únicamente de la distancia temporal $k$ entre ellas, no del momento $t$ en que se miden.
    $$Cov(Z_t, Z_{t+k}) = \gamma_k$$


### 3. La Función de Autocovarianza

La función de autocovarianza mide la dirección y fuerza de la relación lineal entre las observaciones de la serie separadas por $k$ periodos de tiempo (rezagos o *lags*).

Se define formalmente como $\gamma_k$:

$$\gamma_k = Cov(Z_t, Z_{t+k}) = E[(Z_t - \mu)(Z_{t+k} - \mu)]$$

Propiedades importantes:
* **Simetría:** $\gamma_k = \gamma_{-k}$
* **Varianza:** Para el rezago $k=0$, la autocovarianza es igual a la varianza del proceso: $\gamma_0 = \sigma_Z^2$.


### 4. Proceso de Ruido Blanco

El Ruido Blanco se define como un proceso puramente aleatorio, sin memoria y sin estructura de dependencia predecible. En la modelación ARIMA, los residuos o innovaciones del modelo ($a_t$) deben comportarse como ruido blanco para considerar que el modelo es adecuado.

Un proceso $\{a_t\}$ es **Ruido Blanco** si satisface:

1.  **Media Cero:**
    $$E[a_t] = 0$$
    
2.  **Varianza Constante:**
    $$Var(a_t) = \sigma_a^2$$
    
3.  **Incorrelación (Ausencia de patrón):** No existe relación lineal entre observaciones distintas; las autocovarianzas son cero para cualquier rezago $k \neq 0$.
    $$\gamma_k = Cov(a_t, a_{t+k}) = \begin{cases} \sigma_a^2 & \text{si } k = 0 \\ 0 & \text{si } k \neq 0 \end{cases}

Si además de estas propiedades, las variables siguen una distribución normal, se denomina **Ruido Blanco Gaussiano**.

### 5. Procesos Autorregresivos $AR(p)$

Los modelos autorregresivos son los modelos más simples de procesos estacionarios utilizados para representar la dependencia de los valores de una serie de tiempo respecto a su pasado. Generalizan la idea de regresión lineal, donde el valor actual de la serie depende linealmente de sus propios valores anteriores más un término de error.

Una serie de tiempo $\{Z_{t}\}$ estacionaria sigue un proceso autorregresivo de orden $p$ si se puede expresar en términos de sus retardos hasta $z_{t-p}$ y una innovación $a_{t}$:

$$z_{t} = c + \phi_{1}z_{t-1} + \cdot\cdot\cdot + \phi_{p}z_{t-p} + a_{t}$$

Donde:
* $\phi_{1}, ..., \phi_{p}$ son los parámetros autorregresivos.
* $a_{t}$ es un proceso de ruido blanco.
* La condición necesaria para que el proceso sea estacionario es que las raíces de su ecuación característica se encuentren fuera del círculo unitario (sean mayores a 1 en módulo).

### 6. Procesos de Medias Móviles $MA(q)$

Los procesos de medias móviles tienen la propiedad de representar series de tiempo de "memoria corta". En estos modelos, el valor actual de la serie se expresa como una función lineal de un número finito de innovaciones pasadas.

Un proceso de medias móviles de orden $q$, denotado como $MA(q)$, se representa como:

$$\tilde{z}_{t} = a_{t} - \theta_{1}a_{t-1} - \theta_{2}a_{t-2} - \cdot\cdot\cdot - \theta_{q}a_{t-q}$$

Donde:
* $\tilde{z}_{t} = z_{t} - \mu$ (desviación respecto a la media).
* $\theta_{1}, ..., \theta_{q}$ son los parámetros de medias móviles.
* $a_{t}$ son las innovaciones (ruido blanco).

**Propiedades:**
* Un proceso $MA(q)$ siempre es estacionario, ya que es la suma de procesos estacionarios (ruido blanco).
* Se dice que el modelo es **invertible** si las raíces del operador de medias móviles son mayores que uno en módulo.

### 7. Modelo Mixto $ARMA(p,q)$

Los procesos Autorregresivos de Medias Móviles (ARMA) combinan ambas estructuras. Permiten modelar series estacionarias utilizando una parte autorregresiva de orden $p$ (que explica la dependencia geométrica del pasado) y una parte de medias móviles de orden $q$ (que explica la dependencia de las innovaciones recientes).

El modelo $ARMA(p,q)$ se define como:

$$z_{t} = c + \sum_{i=1}^{p}\phi_{i}z_{t-i} + a_{t} - \sum_{j=1}^{q}\theta_{j}a_{t-j}$$

Este modelo es útil porque puede representar procesos complejos con menos parámetros que un modelo AR o MA puro equivalente.

Para que este proceso sea válido se requiere:
* **Estacionariedad:** Determinada por las raíces de la parte autorregresiva ($\phi$).
* **Invertibilidad:** Determinada por las raíces de la parte de medias móviles ($\theta$).
